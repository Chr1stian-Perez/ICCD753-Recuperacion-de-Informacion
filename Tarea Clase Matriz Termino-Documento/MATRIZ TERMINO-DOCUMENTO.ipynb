```python
import math
def crear_matriz_termino_documento(ruta_archivo):
    try:
        with open(ruta_archivo, 'r', encoding='utf-8') as archivo:
            corpus = archivo.readlines()
    except Exception as e:
        print(f"Error al leer el archivo: {e}")
        return None
    N = len(corpus)
    diccionario = set()
    reemplazos = {
        'á': 'a', 'é': 'e', 'í': 'i', 'ó': 'o', 'ú': 'u',
        'Á': 'a', 'É': 'e', 'Í': 'i', 'Ó': 'o', 'Ú': 'u',
        'ü': 'u', 'Ü': 'u', 'ñ': 'n', 'Ñ': 'n'
    }
    puntuacion = '.,;:¡!¿?()[]{}"\'«»-_—+=*/\\|<>@#$%&'
    documentos_procesados = []
    for doc in corpus:
        doc = doc.lower()
        for original, reemplazo in reemplazos.items():
            doc = doc.replace(original, reemplazo)
        for p in puntuacion:
            doc = doc.replace(p, ' ')
        palabras = doc.split()
        documentos_procesados.append(palabras)
        for palabra in palabras:
            if len(palabra) > 0:
                diccionario.add(palabra)
    diccionario = sorted(list(diccionario))
    matriz = []
    for termino in diccionario:
        fila = [termino] 
        documentos_con_termino = 0
        for doc in documentos_procesados:
            frecuencia = doc.count(termino)
            fila.append(frecuencia)
            if frecuencia > 0:
                documentos_con_termino += 1
        #IDF: log(N/dft)
        if documentos_con_termino > 0:
            idf = math.log10(N / documentos_con_termino)
        else:
            idf = 0
        fila.append(round(idf, 4))
        matriz.append(fila)
    return matriz, diccionario, N
def round(num, digits):
    mult = 10 ** digits
    return int(num * mult + 0.5) / mult
def mostrar_matriz(matriz, N):
    encabezado = ["Término"]
    for i in range(N):
        encabezado.append(f"Doc {i+1}")
    encabezado.append("IDF")
    print(" | ".join(encabezado))
    print("-" * (len(encabezado) * 8))
    for fila in matriz:
        print(" | ".join(str(celda) for celda in fila))
def main():
    ruta_archivo = r"C:\Users\Christian\Desktop\EPN\2025-A\Recuperacion de Informacion\ir25a-main\data\01_corpus_turismo.txt"
    resultado = crear_matriz_termino_documento(ruta_archivo)
    if resultado:
        matriz, diccionario, N = resultado
        print(f"Total de documentos: {N}")
        print(f"Total de términos únicos: {len(diccionario)}")
        mostrar_matriz(matriz, N)
if __name__ == "__main__":
    main()
```
